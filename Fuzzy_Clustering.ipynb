```python
from fcmeans import FCM
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
from seaborn import scatterplot as scatter

import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
mall = pd.read_csv('Mall_Customers.csv')
```

```python
mall = mall.drop(['CustomerID','Genre'],axis=1)
```

```python
mall = mall.values
```

#Fuzzy with 2 Clusters

```python
# fit the fuzzy-c-means
fcm = FCM(n_clusters=2)
fcm.fit(mall)

# outputs
fcm_centers = fcm.centers
fcm_labels  = fcm.u.argmax(axis=1)
fcm_predict = fcm.predict(mall)

#Clustering metrics
#davies_bouldin_score
from sklearn.metrics import davies_bouldin_score
clustering_score = davies_bouldin_score(mall, fcm_labels)
print("clustering_score : ",clustering_score)

#silhouette_score
from sklearn.metrics import silhouette_score
silhouette_score = silhouette_score(mall, fcm_labels)
print("silhouette_score : ",silhouette_score)

# plot result
%matplotlib inline
f, axes = plt.subplots(1, 2, figsize=(11,5))
scatter(mall[:,0], mall[:,1], ax=axes[0])
scatter(mall[:,0], mall[:,1], ax=axes[1], hue=fcm_labels)
scatter(fcm_centers[:,0], fcm_centers[:,1], ax=axes[1],marker="s",s=200)
plt.show()
```
