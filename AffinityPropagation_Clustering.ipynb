```
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import AffinityPropagation

```

```
dataset = pd.read_csv('Mall_Customers.csv')
```

```
dataset.head()
```

```
X = dataset.iloc[:, [3, 4]].values
```

```
clustering = AffinityPropagation().fit(X)
```

```
clustering.labels_
```

#Clustering metrics

```
#Clustering metrics
from sklearn.metrics import davies_bouldin_score

clustering_score = davies_bouldin_score(X, clustering.labels_)

print("davies_bouldin clustering_score : ",clustering_score)
```

```
from sklearn.metrics import silhouette_score

silhouette_score = silhouette_score(X, clustering.labels_)

print("silhouette clustering_score : ",silhouette_score)
```
